```python
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

Data = {'x': [25,34,22,27,33,33,31,22,35,34,67,54,57,43,50,57,59,52,65,47,49,48,35,33,44,45,38,43,51,46],
        'y': [79,51,53,78,59,74,73,57,69,75,51,32,40,47,53,36,35,58,59,50,25,20,14,12,20,5,29,27,8,7]
       }
  
df = DataFrame(Data,columns=['x','y'])
  
kmeans = KMeans(n_clusters=3).fit(df)

centroids = kmeans.cluster_centers_
print(centroids)

plt.scatter(df['x'], df['y'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show() 
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd

# 1. Simple transaction data (1 = bought, 0 = not)
# Items: [Milk, Bread, Eggs, Beer]
data = [[1, 1, 0, 0], [0, 1, 1, 0], [1, 1, 1, 0], [0, 0, 0, 1]]
df = pd.DataFrame(data, columns=['Milk', 'Bread', 'Eggs', 'Beer'])

# 2. Find items that appear in at least 50% of carts
frequent_items = apriori(df, min_support=0.5, use_colnames=True)

# 3. Create the rules!
rules = association_rules(frequent_items, metric="confidence", min_threshold=0.7)

print(rules[['antecedents', 'consequents', 'support', 'confidence']])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

movies = ["ghilli","goa","mumbai","vadachennai","bigil"]

user_input = input("Enter movie: ").lower()

# convert text to vectors
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(movies + [user_input])

# similarity
sim = cosine_similarity(vectors[-1], vectors[:-1])

# best match
index = sim.argmax()

score = sim[0][index]

if score > 0.3:
    print("Suggested movie:", movies[index])
else:
    print("No similar movie found")
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# 1. Setup the Matrix (10 Movies x 5 Users)
data = {
    'Inception': [5, 4, 1, 0, 2],
    'The Dark Knight': [4, 5, 2, 1, 1],
    'Interstellar': [5, 5, 3, 2, 2],
    'The Godfather': [2, 1, 5, 4, 5],
    'Pulp Fiction': [1, 2, 4, 5, 4],
    'The Matrix': [5, 4, 2, 1, 3],
    'Forrest Gump': [3, 3, 4, 4, 5],
    'The Shawshank Redemption': [2, 2, 5, 5, 5],
    'Gladiator': [4, 3, 3, 2, 4],
    'Titanic': [3, 2, 2, 5, 4]
}

df = pd.DataFrame(data, index=['User_1', 'User_2', 'User_3', 'User_4', 'User_5'])

# 2. Pre-calculate Matrix Similarity
# Transpose so movies are rows for cosine calculation
movie_matrix = df.T
sim_matrix = cosine_similarity(movie_matrix)
sim_df = pd.DataFrame(sim_matrix, index=df.columns, columns=df.columns)

# 3. Display Available Movies
print("--- Available Movies in Database ---")
available_movies = list(df.columns)
for i, movie in enumerate(available_movies, 1):
    print(f"{i}. {movie}")

# 4. User Interaction
user_input = input("\nEnter the name of a movie you liked: ").strip()

if user_input in available_movies:
    # Get the average rating of the input movie
    avg_rating = df[user_input].mean()
    print(f"\n> You selected: {user_input}")
    print(f"> Current Average User Rating: {avg_rating:.2f}/5.0")

    # 5. Recommendation Logic (Top 3)
    print(f"\n--- Top 3 Recommendations based on {user_input} ---")
    # Sort by similarity, skip the first one (itself)
    recommendations = sim_df[user_input].sort_values(ascending=False)[1:4]
    
    for movie, score in recommendations.items():
        movie_rating = df[movie].mean()
        print(f"Movie: {movie} | Similarity Score: {score:.3f} | Avg Rating: {movie_rating:.2f}")
else:
    print("\nError: Movie not found. Please check your spelling and try again.")
```

```python
from sklearn.ensemble import IsolationForest
import pandas as pd

# 1. Normal heart rates (60-100) vs an Anomaly (200!)
data = [[70], [72], [80], [75], [200], [68]]
df = pd.DataFrame(data, columns=['HeartRate'])

# 2. Setup the "Forest" (contamination is the % of weirdness we expect)
model = IsolationForest(contamination=0.15)

# 3. Fit and Predict
# 1 = Normal, -1 = Anomaly!
df['anomaly'] = model.fit_predict(df[['HeartRate']])

print(df)
```
