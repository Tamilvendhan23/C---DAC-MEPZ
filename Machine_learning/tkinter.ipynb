```python
import pickle
import tkinter as tk
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
def click():
    f1 = open("LRM.pkl", "rb")
    model = pickle.load(f1)
    a = int(e1.get())
    b = int(e2.get())
    c = int(e3.get())
    d = int(e4.get())
    data = np.array([[a, b, c, d]])
    result = model.predict(data)
    f1.close()
    if result[0] == 1:
        e5.insert(0,"YES")
    else:
        e5.insert(0,"NO")
```

```python
root = tk.Tk()
```

```python
l1 = tk.Label(root,text="PLAYGOLF")
l2 = tk.Label(root,text="Outlook")
l3 = tk.Label(root,text="Temp")
l4 = tk.Label(root,text="Humidity")
l5 = tk.Label(root,text="Windy")
l6 = tk.Label(root,text="PlayGolf")
```

```python
e1= tk.Entry(root)
e2= tk.Entry(root)
e3= tk.Entry(root)
e4= tk.Entry(root)
e5= tk.Entry(root)
```

```python
import sys

def fun():
    sys.exit(0)

b1 = tk.Button(root, text = "predict",command=click)
b2 = tk.Button(root, text = "exit")
```

```python
l1.grid(row=0,column=0)
l2.grid(row=1,column=0)
e1.grid(row=1,column=1)
l3.grid(row=2,column=0)
e2.grid(row=2,column=1)
l4.grid(row=3,column=0)
e3.grid(row=3,column=1)
l5.grid(row=4,column=0)
e4.grid(row=4,column=1)
l6.grid(row=5,column=0)
e5.grid(row=5,column=1)
b1.grid(row=6,column=0)
b2.grid(row=6,column=1)
root.mainloop()
```
