## python program to connect database

```python
import sqlite3
try:
    # Step 2: connect to the database 
    con = sqlite3.connect("master.db")

    # Step 3: create a cursor
    cur = con.cursor()

    # Step 4: create a table (if not exists)
    cur.execute("CREATE TABLE IF NOT EXISTS jeevitha (id INTEGER PRIMARY KEY, name TEXT)")

    # Step 5: insert sample data
    cur.execute("INSERT INTO jeevitha (name) VALUES ('AADHI')")
    con.commit()

    # Step 6: update the record (convert name to uppercase)
    cur.execute("UPDATE jeevitha SET name = UPPER(name) WHERE name = 'AADHI'")
    con.commit()

    # Step 7: delete a record
    cur.execute("DELETE FROM jeevitha WHERE name = 'Sample Name'")
    con.commit()

    # Step 8: execute select query
    cur.execute("SELECT * FROM jeevitha")

    # Step 9: display results
    for elem in cur.fetchall():
        print(elem)
except sqlite3.Error as e1:
    print("SQL Error = ", e1)
except Exception as e2:
    print("Other Error = ", e2)
finally:
    # Step 10: close cursor and connection
    if cur:
        cur.close()
    if con:
        con.close()
    
```
