```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate dates for March 2026
dates = pd.date_range(start="2026-03-01", end="2026-03-31")

# Simulated temperature data
np.random.seed(0)
temperature = np.random.uniform(29, 33, len(dates)).round(2)

# Build DataFrame
df = pd.DataFrame({"Date": dates, "Temperature": temperature})
df.set_index("Date", inplace=True)

# --- Inject anomaly values for 15–20 March ---
df.loc["2026-03-15":"2026-03-20", "Temperature"] = [45, 44, 43, 42, 41, 40]
df["Moving_Avg"] = df["Temperature"].rolling(window=3).mean()

# Residual = Actual - Moving Average
df["Residual"] = df["Temperature"] - df["Moving_Avg"]

print(df)

from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose with daily frequency (period=7 for weekly seasonality)
result = seasonal_decompose(df["Temperature"], model="additive", period=7)


result.plot()
plt.show()
```

```python
import pandas as pd
from datetime import date
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.stattools import adfuller

a = [date(2024,1,1),date(2024,1,2),date(2024,1,3),date(2024,1,4),
     date(2024,1,5),date(2024,1,6),date(2024,1,7),date(2024,1,8)]

b = [10,8,10,8,10,8,10,8]

df = pd.DataFrame({"dob":a, "temp":b})

df["dob"] = pd.to_datetime(df["dob"])
df.set_index("dob", inplace=True)

# stationarity test
result = adfuller(df['temp'])
print("P-value:", result[1])

# plot
df["temp"].plot()
plt.show()

# AutoReg model
ar_model = AutoReg(df["temp"], lags=1).fit()

# predict next value
pred = ar_model.predict(start=len(df), end=len(df))
print("Next prediction:", pred)
```

```python
from datetime import date,timedelta
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
from statsmodels.tsa.stattools import adfuller

a=[]
for _ in range(0,100):
  a.append(date(2024,1,1)+timedelta(_))

b=[]
for _ in range(0,100):
   b.append(random.randrange(23,40))


df = pd.DataFrame({"date":a, "temp":b})

df["date"] = pd.to_datetime(df["date"])

# set the row index as date
df.index = df["date"]
df.drop(["date"], axis=1, inplace=True)

# log and diff
df["temp_log"]  = np.log(df["temp"])
df["temp_diff_log"] = df["temp_log"].diff()

def adf_test(series):
   test_results = adfuller(series)
   print("ADF statistics = ",test_results[0])
   print("P-Value        = ",test_results[1])
   print("Critical values:")
   for thres,adf_stat in test_results[4].items():
       print("\t%s: %.2f" %(thres,adf_stat))

adf_test(df["temp_diff_log"][1:])
```
