```python
df = pd.read_csv("file12.csv")
df
```

```python
df["xy"] = df["x"]*df["y"]
df
```

```python
df["x^2"] = df["x"]**2
df
```

```python
x_mean = (df["x"].sum())/len(df["x"])
x_mean
```

```python
y_mean = (df["y"].sum())/len(df["y"])
y_mean
```

```python
x_sqr = (df["x^2"].sum())/len(df["x^2"])
x_sqr
```

```python
xy = df["xy"].mean()
xy
```

```python
t = x_mean**2
t
```

```python
res = ((xy)-(x_mean*y_mean))/((x_sqr)-(t))
print(res)
```
