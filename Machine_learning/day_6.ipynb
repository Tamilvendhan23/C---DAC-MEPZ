```python
import pandas as pd

f = open("data69.csv","w")
f.write("""x,y
-9,0
-8,0
0,1
8,1
9,1""")
f.close()

df = pd.read_csv("data69.csv")

x_mean = df["x"].mean()
y_mean = df["y"].mean()

df["xy"] = df["x"] * df["y"]
xy = df["xy"].mean()

df["x2"] = df["x"]**2
sqr = df["x2"].mean()

# slope
b1 = (xy - (x_mean*y_mean)) / (sqr - (x_mean**2))

# intercept
a0 = y_mean - b1*x_mean
df
print("slope:", b1)
print("intercept:", a0)
df
```
