```python
import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('image.jpeg')

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Show original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct colors
plt.title("Original Image")
plt.axis("off")

# Show grayscale image
plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='grey')
plt.title("Grayscale Image")
plt.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('image.jpeg')

# Resize the image (width=400, height=300)
resized_image = cv2.resize(image, (300, 200))

# Print (display) the resized image inline
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))  # Convert BGR → RGB
plt.title("Resized Image")
plt.axis("off")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread('image.jpeg')

# check if loaded
if image is None:
    print("Image not found")
    exit()

# BEFORE size
print("Original size:", image.shape)   # (height, width, channels)

# Resize (example: 300x300)
resized_image = cv2.resize(image, (100, 100))

# AFTER size
print("Resized size:", resized_image.shape)

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_resized = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

# Show images
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title("Resized")
plt.axis("off")

plt.show()
```
