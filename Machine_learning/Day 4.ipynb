```python
import pandas as pd

f = open("data47.csv","w")  # if i using diffrenct dir need to specify liek (r"c:\\that\\this\\data.csv") here r is raw string
f.write("""num1,num2
10,20
30,40
50,a
60,70
80,b
90,15
c,25
10,15""")
f.close()
df = pd.read_csv("data47.csv")

df = df.replace(["a","b","c"], 0)
df["num1"]=df["num1"].astype(int)
df["num2"]=df["num2"].astype(int)


df["sum"] = df["num1"] + df["num2"]
df["avg"] = df["sum"] / 2

print(df)
df.to_csv("out47.csv")
```

```python
import pandas as pd

f = open("data50.csv","w")
f.write("""name  dept
ravi  sales
manu  purch
hari  accts
ranjan  finan""")
f.close()

df = pd.read_csv("data50.csv", sep="\s+")

df1 = df.copy()
df1["name"] = df1["name"].apply(lambda x: x[0].upper() + x[1:-1] + x[-1].upper())
print(df1)

# Expected2 (all upper case)

df2 = df.copy()
df2["name"] = df2["name"].str.upper()
df2["dept"] = df2["dept"].str.upper()

print(df2)
```
