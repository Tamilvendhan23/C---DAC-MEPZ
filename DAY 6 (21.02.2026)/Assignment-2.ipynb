# Ex2 - Filtering and Sorting Data


### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](./chipotle.tsv). 

```python
chipo = pd.read_csv("chipotle.tsv", sep="\t")
```

### Step 3. Assign it to a variable called chipo.

```python
chipo["item_price"] = chipo["item_price"].apply(lambda x: float(x.replace("$", "")))
```

### Step 4. How many products cost more than $10.00?

```python
chipo[chipo["item_price"] > 10]["item_name"].nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo[["item_name", "item_price"]].drop_duplicates()
```

### Step 6. Sort by the name of the item

```python
chipo[["item_name", "item_price"]].drop_duplicates().sort_values("item_name")
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo[chipo["item_price"] == chipo["item_price"].max()]["quantity"].sum()
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
chipo[chipo["item_name"] == "Veggie Salad Bowl"]["quantity"].sum()
```

### Step 9. How many times people orderd more than one Canned Soda?

```python
chipo[
    (chipo["item_name"] == "Canned Soda") & (chipo["quantity"] > 1)
]["quantity"].count()
```
