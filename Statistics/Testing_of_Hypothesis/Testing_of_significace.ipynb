# Q15 - Two Sample t-Test

```python
import math

# Given samples
sample1 = [9, 11, 13, 11, 15, 9, 12, 14]
sample2 = [10, 12, 10, 14, 9, 8, 10]

# Step 1: Length of samples
n1 = len(sample1)
n2 = len(sample2)

# Step 2: Calculate mean manually
mean1 = sum(sample1) / n1
mean2 = sum(sample2) / n2

# Step 3: Calculate variance manually
var1 = sum((x - mean1)**2 for x in sample1) / (n1 - 1)
var2 = sum((x - mean2)**2 for x in sample2) / (n2 - 1)

# Step 4: Pooled variance
df = n1 + n2 - 2
pooled_var = ((n1-1)*var1 + (n2-1)*var2) / df

# Step 5: Standard error
se = math.sqrt(pooled_var * (1/n1 + 1/n2))

# Step 6: t statistic
t_value = (mean1 - mean2) / se

print("Mean 1 =", mean1)
print("Mean 2 =", mean2)
print("Pooled Variance =", pooled_var)
print("t statistic =", t_value)

print("If |t| is small, difference is not significant")
```

# Q16 - Chi Square Test (from scratch)

```python
import math

# Observed values
observed = [12, 40, 88, 110, 56, 14]

total = 320
n = 5
p = 0.5

# Step 1: Calculate expected frequencies using Binomial formula
expected = []

for k in range(n+1):
    combination = math.comb(n, k)
    probability = combination * (p**k) * ((1-p)**(n-k))
    exp_value = probability * total
    expected.append(exp_value)

print("Expected Frequencies:", expected)

# Step 2: Calculate Chi-square manually
chi_square = 0

for o, e in zip(observed, expected):
    chi_square += ((o - e)**2) / e

print("Chi Square value =", chi_square)

print("If Chi-square value is small, hypothesis is accepted")
```

# Q17 - One Way ANOVA

```python
# Given data
A = [36, 37, 42, 38, 47]
B = [46, 39, 35, 37, 43]
C = [35, 42, 37, 43, 38]
D = [45, 36, 39, 35, 32]
E = [41, 39, 37, 35, 38]

groups = [A, B, C, D, E]

# Step 1: Find grand mean
all_values = []

for group in groups:
    for value in group:
        all_values.append(value)

grand_mean = sum(all_values) / len(all_values)

# Step 2: Calculate SSB
SSB = 0

for group in groups:
    group_mean = sum(group) / len(group)
    SSB += len(group) * (group_mean - grand_mean)**2

# Step 3: Calculate SSW
SSW = 0

for group in groups:
    group_mean = sum(group) / len(group)
    for value in group:
        SSW += (value - group_mean)**2

# Step 4: Mean squares
df_between = len(groups) - 1
df_within = len(all_values) - len(groups)

MSB = SSB / df_between
MSW = SSW / df_within

F_value = MSB / MSW

print("Grand Mean =", grand_mean)
print("SSB =", SSB)
print("SSW =", SSW)
print("F statistic =", F_value)

print("If F value is small, no significant difference")
```
