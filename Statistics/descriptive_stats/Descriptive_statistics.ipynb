#QUESTION 2 – Python Statistical Implementation

```python
import numpy as np

# Class intervals
classes = [(12,17),(17,22),(22,27),(27,32),
           (32,37),(37,42),(42,47),(47,52),(52,57)]

# Frequencies
frequency = np.array([6,7,5,8,18,15,11,14,4])

# Step 1: Find midpoints
midpoints = np.array([(a+b)/2 for a,b in classes])

# Step 2: Total frequency
N = np.sum(frequency)

# Step 3: Mean
mean = np.sum(midpoints * frequency) / N

# Step 4: Median (Grouped formula)
cumulative_freq = np.cumsum(frequency)
median_position = N / 2

median_class_index = np.where(cumulative_freq >= median_position)[0][0]

L = classes[median_class_index][0]
cf_previous = cumulative_freq[median_class_index - 1]
f = frequency[median_class_index]
h = classes[median_class_index][1] - classes[median_class_index][0]

median = L + ((median_position - cf_previous) / f) * h

# Step 5: Mode (Grouped formula)
modal_class_index = np.argmax(frequency)

L_mode = classes[modal_class_index][0]
f1 = frequency[modal_class_index]
f0 = frequency[modal_class_index - 1]
f2 = frequency[modal_class_index + 1]
h_mode = classes[modal_class_index][1] - classes[modal_class_index][0]

mode = L_mode + ((f1 - f0) / (2*f1 - f0 - f2)) * h_mode

print("Mean =", mean)
print("Median =", median)
print("Mode =", mode)
```

#QUESTION 4 – Python Code Using SciPy

```python
import numpy as np
from scipy import stats

# Class intervals
classes = [(5,10),(10,15),(15,20),(20,25),
           (25,30),(30,35),(35,40),(40,45)]

# Frequencies
frequency = np.array([3,5,7,6,12,10,11,7])

# Midpoints
midpoints = np.array([(a+b)/2 for a,b in classes])

# Expand grouped data
data = np.repeat(midpoints, frequency)

# Mean using SciPy
mean = stats.tmean(data)

# Standard deviation using SciPy
std_dev = stats.tstd(data)   # sample standard deviation

print("Mean =", mean)
print("Standard Deviation =", std_dev)
```
