#Qn-18 : Correlation and Regression
#Verification using scikit-learn & scipy

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr

# Data
X = np.array([35, 66, 78, 75, 74, 49, 62, 59, 66, 78]).reshape(-1,1)
Y = np.array([48, 62, 71, 79, 63, 42, 68, 53, 67, 85])

print("---- Verification using sklearn ----")

# 1. Correlation
r_value, p_value = pearsonr(X.flatten(), Y)
print("Correlation (r) =", round(r_value,4))

# 2. Linear Regression
model = LinearRegression()
model.fit(X, Y)

print("Intercept (a) =", round(model.intercept_,4))
print("Slope (b) =", round(model.coef_[0],4))

print("Regression Equation:")
print("Y =", round(model.intercept_,2), "+", round(model.coef_[0],2), "X")
```

```python
# Q18 - Correlation and Regression (FULL SCRATCH MODE)

import math

# Given Data
X = [35, 66, 78, 75, 74, 49, 62, 59, 66, 78]
Y = [48, 62, 71, 79, 63, 42, 68, 53, 67, 85]

n = len(X)

# -----------------------------
# STEP 1: Basic Summations
# -----------------------------

sum_x = sum(X)
sum_y = sum(Y)

sum_x2 = sum(x**2 for x in X)
sum_y2 = sum(y**2 for y in Y)
sum_xy = sum(X[i] * Y[i] for i in range(n))

print("ΣX =", sum_x)
print("ΣY =", sum_y)
print("ΣX² =", sum_x2)
print("ΣY² =", sum_y2)
print("ΣXY =", sum_xy)

# -----------------------------
# STEP 2: Correlation Formula
# r = [nΣXY - ΣXΣY] /
#     sqrt[(nΣX² - (ΣX)²)(nΣY² - (ΣY)²)]
# -----------------------------

numerator = (n * sum_xy) - (sum_x * sum_y)

denominator_part1 = (n * sum_x2) - (sum_x ** 2)
denominator_part2 = (n * sum_y2) - (sum_y ** 2)

denominator = math.sqrt(denominator_part1 * denominator_part2)

r = numerator / denominator

print("\nCorrelation Coefficient (r) =", r)

# -----------------------------
# STEP 3: Regression Coefficients
# b = numerator / denominator_part1
# a = (ΣY - bΣX) / n
# -----------------------------

b = numerator / denominator_part1
a = (sum_y - b * sum_x) / n

print("\nSlope (b) =", b)
print("Intercept (a) =", a)

print("\nRegression Equation:")
print("Y =", a, "+", b, "X")

# -----------------------------
# STEP 4: Interpretation
# -----------------------------

if r > 0.8:
    print("\nStrong positive correlation exists.")
else:
    print("\nModerate correlation exists.")

print("For every 1 mark increase in Programming,")
print("Control System marks increase by approximately", b)
```
