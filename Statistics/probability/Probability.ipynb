# Question 5
# Probability that A, B, C solve the problem

```python
P_A = 1/2
P_B = 1/3
P_C = 1/4

# Step 1: Probability that none solve
P_not_A = 1 - P_A
P_not_B = 1 - P_B
P_not_C = 1 - P_C

P_none_solve = P_not_A * P_not_B * P_not_C

# Step 2: Probability that problem is solved
P_solved = 1 - P_none_solve

print("Probability that none solve =", P_none_solve)
print("Probability that problem is solved =", P_solved)
```

# Question (9 Bayes Theorem)
# Ratio of becoming GM: 4 : 2 : 3

```python
# Step 1: Convert ratio into probabilities
total = 4 + 2 + 3

P_A = 4 / total
P_B = 2 / total
P_C = 3 / total

# Given conditional probabilities
P_D_given_A = 0.3
P_D_given_B = 0.7
P_D_given_C = 0.8

# Step 2: Total probability of D
P_D = (
    P_A * P_D_given_A +
    P_B * P_D_given_B +
    P_C * P_D_given_C
)

# Step 3: Bayes theorem
P_A_given_D = (P_A * P_D_given_A) / P_D

print("Total Probability P(D) =", P_D)
print("Probability A is GM given D =", P_A_given_D)
```
