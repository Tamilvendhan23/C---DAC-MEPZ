#Question 12 - (Defective bolts – Binomial Distribution)

```python
import math

n = 15
p = 0.05

def binomial(r):
    return math.comb(n, r) * (p**r) * ((1-p)**(n-r))

# Exactly 3 defective
prob_3 = binomial(3)
print("Probability of exactly 3 defective =", prob_3)

# Not more than 3 defective
prob_upto3 = sum(binomial(r) for r in range(4))
print("Probability of not more than 3 defective =", prob_upto3)
```

#Q14 — NORMAL DISTRIBUTION

```python
  from scipy.stats import norm

mean = 50
std = 18
total = 2000

# (i) Over 90
prob_over90 = 1 - norm.cdf(90, mean, std)
print("Children scoring above 90 =", total * prob_over90)

# (ii) Below 40
prob_below40 = norm.cdf(40, mean, std)
print("Children scoring below 40 =", total * prob_below40)

# (iii) Between 50 and 80
prob_between = norm.cdf(80, mean, std) - norm.cdf(50, mean, std)
print("Children scoring between 50 and 80 =", total * prob_between)
```
